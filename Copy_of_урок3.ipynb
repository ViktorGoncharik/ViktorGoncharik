#Вектор

Вектором называется направленный отрезок, для которого указано его начало и конец
image.png

1) Векторы можно записать двумя большими латинскими буквами:
 и так далее. При этом первая буква обязательно обозначает точку-начало вектора, а вторая буква – точку-конец вектора.

2) Векторы также записывают маленькими латинскими буквами:
 В частности, наш вектор  можно для краткости переобозначить маленькой латинской буквой .

Длиной или модулем ненулевого вектора  называется длина отрезка . Длина нулевого вектора  равна нулю. Логично.

Длина вектора обозначается знаком модуля: , | |

```python
plt.plot([0,10],[0,8])
# (0,10)(0,8) (10)(8)
```

```python
plt.xlabel("Ось Х")
plt.ylabel("Ось Y")
plt.title("Вектор на плоскости")
plt.plot([8,8,8],[0,5,10],label = 'x1')
plt.plot([0,4,8],[10,10,10],label = 'x2')


plt.annotate('', xy=(8,10),  xycoords='data',
            xytext=(0, 0), textcoords='data',
            arrowprops=dict(facecolor='blue'))

plt.text(3, 3, "Вектор", fontsize=20)
plt.grid(True)
plt.legend()
plt.show()
```

Сгенерерируй случайную выборку из 31 значения. И построй две гистограммы с четными и нечетными индексами

```python
a = np.random.randint(1,100,31)
even = []
odd = []
for each in a:
    if each%2==0:
      even.append(each)
    else:
      odd.append(each)

print(even)
print(odd)

plt.hist(even)
plt.hist(odd)

```
